```python
#01: 
class Student:
    def __init__(self,name,marks,gender):
        self.name=name
        self.marks=marks
        self.gender=gender
    def display_info(self):
        print(f'name:{self.name} \ngrade :{self.marks}\ngender:{self.gender}') 
s1=Student("noshabah",85,'f')   #object 
s1.display_info()  
```

```python
#02:
class Car:
    def __init__(self,color):
        self.color=color

    def start(self):
        print(self.color, "car started") 
car1=Car("red")    
car1.start()
car2=Car("white")    
car2.start()
```

```python
#03:
#INHERITENCE:

class Animal:
    def speak(self):
        print("some sound...")
class Dog(Animal): #inheritence
    def bark(self):
        print("woof!")
dog1=Dog()     
dog1.speak()  # call to method that it inheret
dog1.bark()  #call to its own method

class Cat(Animal):
    def speak(self):
        print("meow meow!!") #method over riding...
cat1=Cat()   
cat1.speak()     


#So here the Dog clas inheret the Animal class and is able to use its properties and methods..
```

```python
#04:
#Multiple Inheretance

class A:
    varA="welcome to class A"
class B:
    varB="welcome to class B" 
class C(A,B):    #child class
    varC="welcome to class C"   
c1=C()     #inheret properties of both classes...
print(c1.varA)
print(c1.varB)
print(c1.varC)
```

```python
#05:
#multi level inheretance:

class Car:
    def __init__(self,brand):
        self.brand=brand
    def info(self):
        print("car's brand is",self.brand)
    def start(self) :
        print("car has started...") 
class Civic(Car):
    def __init__(self,brand):
        super().__init__(brand)  #used to access parent methods...
c1=Civic("civic")    
print(c1.brand)
c1.start()
c1.info()
class Toyota(Civic):
    def __init__(self,brand):
        super().__init__(brand)
c2=Toyota("toyota")
print(c2.brand)
c2.info()      
c2.start()
```

```python
#06: 
#Bank Account Code

class Account:
    def __init__(self,blnc,acc):
        self.blnc=blnc
        self.acc=acc
    def debit(self):
        debit_amount=int(input("enter money to debit:"))
        if debit_amount<0:
            print("Invalid Amount!")
        elif debit_amount>self.blnc:
            print("Insufficient Balance!")
        else:
            self.blnc-=debit_amount
            print("current balance after debit:",self.blnc)
    def credit(self):
        credit_amount=int(input('enter money to credit:'))
        if credit_amount<0:
            print("Invalid Amount!")
        else:    
            self.blnc+=credit_amount
            print("your new balance after credit :",self.blnc)
acc1=Account(25000,1234567890)  
print("your current balance is:",acc1.blnc,"account is",acc1.acc)     
acc1.debit()

```
