```python
class User:
    def __init__(self,name):
        self.name=name
    def __str__(self) :
        return self.name
class Message:
    def __init__(self,user,textmsg):
        self.user=user
        self.textmsg=textmsg
    def __str__(self):
        return f"{self.user} :{self.textmsg}"    
class ChatRoom:
    def __init__(self,chatroom):
        self.chatroom=chatroom
        self.members=[]      
        self.history=[]
    def sendmessage(self,user,textmsg) :
        msg=Message(user,textmsg)  
        self.history.append(msg)
        print(msg)
    def join(self,user) :
        self.members.append(user)
        print(user,"joined the chat")
    def leave(self,user):
        self.members.remove(user)    
        print(user,"left the chat")
    def chatHistory(self):
        print("\n chat history:\n")
        for msg in self.history:
            print(msg)
fatimah =User("fatimah")   
ahmad=User("ahmad")  
ali=User("ali")      
room=ChatRoom("group chat")
room.join(fatimah)
room.join(ahmad)
room.sendmessage(fatimah,"hello everyone")
room.sendmessage(ahmad,"helloo")
room.join(ali)
room.sendmessage(ali,"what's happening gyzz")
room.leave(fatimah)
room.chatHistory()
```
