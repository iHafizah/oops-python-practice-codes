```python
class Vector:
    def __init__(self,x,y):
    #initializing the vectors with x and y components
        self.x=x
        self.y=y 
    #method to display vectors
    def display_vector(self):
        print(f'Vector({self.x} , {self.y})')
v1=Vector(2,3)  # first vector object
v2=Vector(4,5)  #second vector object
v1.display_vector()
v2.display_vector()
```

```python
# Vector Modeling using OOP in Python

class Vector:
    def __init__(self, x, y):
    #initializing the vectors wuth x and y components
        self.x = x
        self.y = y

    #Add two vectors and return the resultant vector
    def add(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def display(self):
        print(f"Vector({self.x}, {self.y})")

v1 = Vector(3, 4) #first vector (object)
v2 = Vector(2, 5) #second vector (object)

result = v1.add(v2)

#displaying the vectors
v1.display()
v2.display()
print("\nResultant Vector (v1+v2):")
result.display()

```

```python
class Circle:   
    def __init__(self, radius):
        self.radius = radius #initializing the radius of the circle

    #method to calculate the area of the circle
    def calculate_area(self):
        area = 3.14159 * self.radius ** 2
        return area

c1 = Circle(9)  #circle object with radius 9

result = c1.calculate_area() #calculating area.
print("Area of Circle:", result)
```
